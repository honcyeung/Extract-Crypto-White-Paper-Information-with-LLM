```python
import os
import json
```

```python
path = "whitepapers"
list_of_pdfs = []
extracted_info = dict()

for file in os.listdir(path):
    if file.endswith(".pdf"):
        list_of_pdfs.append(file)

# loop through the list of PDFs and process each one with LLM
list_of_pdfs = list_of_pdfs[:2]
for pdf in list_of_pdfs:
    pdf_file = os.path.join(path, pdf)

    print(f"------------ Starting worker for {pdf} ------------")
    try:
        %store pdf_file
        %run llm.ipynb
        print("------------ Worker finished ------------\n")
        %store -r final_output
        extracted_info[pdf] = final_output
    except Exception as e:
        print(f"Error processing {pdf}: {e}")
        # logging?
        continue
```

```python
with open("extracted_info.json", "w") as f:
    json.dump(extracted_info, f, indent = 4)
```
